```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/online_retail_II.csv')
```

### Limpeza Inicial

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

34335 linhas duplicadas removidas

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df['Invoice'].str.startswith("C").sum()
# me trás todo os registros de faturas canceladas
```

```python
df = df[~df['Invoice'].str.startswith("C")]
# Remove todas as faturas canceladas
```

```python
df['Invoice'].str.startswith("C").sum()
```

```python
df.shape
```

Remoção de dados da coluna "Quantity" inválidos

```python
quantity_invalidos = df['Quantity'] < 0
```

```python
quantity_invalidos.sum()
```

```python
print(df.loc[quantity_invalidos, ['Invoice', 'Quantity']].head())
```

```python
df[df['Quantity'] < 0]['Customer ID'].isna().sum()
```

```python
(df['Quantity'] < 0).sum()
```

```python
df = df[~(df['Quantity'] < 0)]
```

```python
(df['Quantity'] < 0).sum()
# Verificação
```

Remoção de dados da coluna "Price" inválidos

```python
df.shape
```

```python
price_invalidos = df['Price'] < 0
price_invalidos.sum()
```

```python
df[df['Price'] < 0]['Customer ID'].isna().sum()
```

```python
df[df['Price'] < 0]
```

```python
df = df[~(df['Price'] < 0)]
```

```python
df.shape
```

Correção dos tipos das colunas

```python
df.dtypes
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['Customer ID'].isna().sum()
```

```python
df['Customer ID'].isna().mean() * 100
```

##### 22,86% da base de dados de Customer ID são valores nulos

```python
df.isnull().sum()
```

```python
df[df['Description'].isna()]
```

```python
# Remoção dos dados nulos de Customer ID
df = df.dropna(subset=['Customer ID'])
```

```python
df['Customer ID'] = df['Customer ID'].astype(int)
```

```python
df.dtypes
```

```python
print(f"Shape inicial: {df_original_shape}")
print(f"Shape final após limpeza: {df.shape}")
```

```python
df.to_csv('../data/processed/online_retail_cleaned.csv', index=False)
print("O Dataset limpo foi salvo com sucesso")
```
