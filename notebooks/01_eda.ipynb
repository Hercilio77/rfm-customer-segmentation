# 01 — Exploração dos Dados (EDA)

## Contexto

Este notebook realiza a exploração inicial do dataset **Online Retail II (UCI)**, 
que contém transações de um e-commerce do Reino Unido entre 2009 e 2011.

O objetivo é entender a estrutura dos dados, identificar problemas de qualidade 
(nulos, duplicatas, inconsistências) e levantar hipóteses para a etapa de limpeza.

**Dataset:** Online Retail II — Year 2009-2010  
**Fonte:** UCI Machine Learning Repository / Kaggle  
**Ferramenta:** Python 3 | Pandas | Matplotlib | Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Configurações visuais

```python
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)
```

#### Carregamento dos Dados

```python
df_original = pd.read_csv('../data/raw/online_retail_II.csv')
```

```python
# Criação da cópia do dataset original
df = df_original.copy()
```

## 2. Visão Geral do Dataset

```python
df.head()
```

```python
print(f"Linhas: {df.shape[0]:,}")
print(f"Colunas {df.shape[1]}")
print(f"\nColunas: {df.columns.tolist()}")
print(f"\nTipos de Dados: ")
print(df.dtypes)
```

## 3. Estatísticas Descritivas

```python
df['Quantity'].describe()
```

```python
df['Price'].describe()
```

```python
print(f"Período: {df['InvoiceDate'].min()} até {df['InvoiceDate'].max()}")
print(f"Clientes únicos: {df['Customer ID'].nunique():,}")
print(f"Produtos únicos: {df['StockCode'].nunique():,}")
print(f"Faturas únicas: {df['Invoice'].nunique():,}")
```

#### 5,9 mil clientes = média de 9 compras por cliente no período registrado.

## 4. Valores Nulos

```python
print("Quantidade de valores nulos:")

print(f"Invoice: {df['Invoice'].isnull().sum()}")
print(f"StockCode: {df['StockCode'].isnull().sum()}")
print(f"Description: {df['Description'].isnull().sum()}")
print(f"Quantity: {df['Quantity'].isnull().sum()}")
print(f"InvoiceDate: {df['InvoiceDate'].isnull().sum()}")
print(f"Price: {df['Price'].isnull().sum()}")
print(f"Customer ID: {df['Customer ID'].isnull().sum()}")
print(f"Country: {df['Country'].isnull().sum()}")
```

## 5. Duplicatas

```python
df.duplicated().sum()
```

## 6. Análise das colunas-chave

```python
cancelamentos = df[df['Invoice'].astype(str).str.startswith('C')]
print(f"Cancelamentos: {len(cancelamentos):,}")
print(f"% do total: {len(cancelamentos)/len(df)*100:.2}%")
```

```python
print(df['Country'].value_counts().head(10))
```

## Conclusões da EDA

Dados com boa variação para RFM (~9 compras/cliente, 5.942 clientes únicos).
Principais problemas a tratar na limpeza: nulos em Customer ID (22%),
duplicatas (34k), cancelamentos (prefixo C), Quantity e Price negativos,
e tipagem incorreta em InvoiceDate e Customer ID.
